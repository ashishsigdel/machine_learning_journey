```python
import pandas as pd 
import numpy as np 
df = pd.read_csv('dc_weather.csv')
df.head()
```

```python
df.info()
```

```python
df1 = df.drop(['name', 'feelslikemax', 'feelslikemin', 'feelslike', 'description', 'icon', 'stations'], axis=1)
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.preciptype = df1.preciptype.fillna('none')
```

```python
df1.isnull().sum()
```

```python
df1.severerisk.isnull().sum()
```

```python
df1['datetime'] = pd.to_datetime(df1['datetime'], errors='coerce')
```

```python
df['datetime'].isna().sum()
```

```python
df1['year'] = df1['datetime'].dt.year
df1['month'] = df1['datetime'].dt.month
df1['day'] = df1['datetime'].dt.day
```

```python
df1.drop(columns=['datetime'], inplace=True)
```

```python
df1.preciptype.unique()
```

```python
df1['preciptype'] = df1['preciptype'].replace('none', '')
preciptype_dummies = df1['preciptype'].str.get_dummies(sep=',')
preciptype_dummies.rename(columns={'freezingrain': 'preci_freezingrain', 'ice': 'preci_ice', 'rain': 'preci_rain', 'snow': 'preci_snow'}, inplace=True)
```

```python
preciptype_dummies.head()
```

```python
df1 = pd.concat([df1, preciptype_dummies], axis=1)
df1 = df1.drop(['preciptype'], axis=1)
df1.head()
```

# using k-nearest for filling nan values for serverrisk

```python
train_data = df1[df1['severerisk'].notna()]
predict_data = df1[df1['severerisk'].isna()]
```

```python
train_data.head()
```

```python
pd.options.display.max_columns = None
predict_data.head()
```

```python
X_train_severerisk = train_data.drop(['sunrise', 'sunset', 'moonphase', 'conditions', 'year', 'month', 'day', 'severerisk'], axis=1)
y_train_severerisk = train_data.severerisk
```

```python
X_train_severerisk.head(3)
```

```python
y_train_severerisk.head(3)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python
param_grid = {'n_neighbors': range(1, 20)}
grid_search = GridSearchCV(KNeighborsRegressor(), param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train_severerisk, y_train_severerisk)
best_n_neighbors = grid_search.best_params_['n_neighbors']
print(f"Optimal n_neighbors: {best_n_neighbors}")
```

```python
best_knn_model = KNeighborsRegressor(n_neighbors=8)
best_knn_model.fit(X_train_severerisk, y_train_severerisk)
predict_data
```

```python
X_predict_severerisk = predict_data.drop(['sunrise', 'sunset', 'moonphase', 'conditions', 'year', 'month', 'day', 'severerisk'], axis=1)
```

```python
X_predict_severerisk.head()
```

```python
predict_data.severerisk = best_knn_model.predict(X_predict_severerisk)
predict_data.severerisk.head(10)
```

```python
df2 = pd.concat([train_data, predict_data], axis=0)
```

```python
df2['sunrise'] = pd.to_datetime(df2['sunrise'])
df2['sunset'] = pd.to_datetime(df2['sunset'])
df2['hour_sunrise'] = df2['sunrise'].dt.hour
df2['min_sunrise'] = df2['sunrise'].dt.minute
df2['sec_sunrise'] = df2['sunrise'].dt.second
df2['hour_sunset'] = df2['sunset'].dt.hour
df2['min_sunset'] = df2['sunset'].dt.minute
df2['sec_sunset'] = df2['sunset'].dt.second
df2.drop(['sunrise', 'sunset'], axis=1, inplace=True)
```

```python
df2.head(3)
```

```python
df2.conditions.unique()
```

```python
unique_conditions = set()
df2['conditions'].str.split(', ').apply(unique_conditions.update)

# Step 2: Create a column for each unique condition, and encode with 1 or 0
for condition in unique_conditions:
    df2[f'condition_{condition}'] = df2['conditions'].apply(lambda x: 1 if condition in x else 0)

# Drop original conditions column if no longer needed
df2.drop('conditions', axis=1, inplace=True)
```

# for sunset and sunrise

```python
y_sunrise = df2[['hour_sunrise', 'min_sunrise', 'sec_sunrise']]
y_sunset = df2[['hour_sunset', 'min_sunset', 'sec_sunset']]
```

```python
X = df2[['year', 'month', 'day', 'cloudcover', 'visibility', 'solarradiation', 
         'solarenergy', 'uvindex', 'moonphase']]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

# Splitting data for sunrise and sunset separately
X_train, X_test, y_train_sunrise, y_test_sunrise = train_test_split(X, y_sunrise, test_size=0.2, random_state=42)
X_train, X_test, y_train_sunset, y_test_sunset = train_test_split(X, y_sunset, test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
# Sunrise model
sunrise_model = RandomForestRegressor(n_estimators=100, random_state=42)
sunrise_model.fit(X_train, y_train_sunrise)
y_pred_sunrise = sunrise_model.predict(X_test)
mae_sunrise = mean_absolute_error(y_test_sunrise, y_pred_sunrise)
print("Sunrise MAE:", mae_sunrise)


# Sunset model
sunset_model = RandomForestRegressor(n_estimators=100, random_state=42)
sunset_model.fit(X_train, y_train_sunset)
y_pred_sunset = sunset_model.predict(X_test)
mae_sunset = mean_absolute_error(y_test_sunset, y_pred_sunset)
print("Sunset MAE:", mae_sunset)
```

```python
sunrise_pred = sunrise_model.predict([[2024, 10, 28, 45, 14, 56, 5, 9,0.33]])
sunset_pred = sunset_model.predict([[2024, 10, 28, 45, 14, 56, 5, 9,0.33]])

# Convert to integers by rounding each element
sunrise_pred_int = np.round(sunrise_pred).astype(int)
sunset_pred_int = np.round(sunset_pred).astype(int)

print("Sunrise prediction (rounded to nearest whole number):", sunrise_pred_int)
print("Sunset prediction (rounded to nearest whole number):", sunset_pred_int)
```

```python
pickle.dump(sunrise_model, open('sunrise_model.pkl', 'wb'))
```

```python
pickle.dump(sunset_model, open('sunset_model.pkl', 'wb'))
```
