```python
no = [11, 29, 44, 55, 47, 28 ,58, 11, 57, 11, 17, 54, 27, 30, 59, 38, 36, 12, 46, 49, 10 ,34, 24, 42,
 35, 33, 51,14, 45, 22, 30 ,17, 38, 34, 36 ,24, 37, 48, 23, 37, 39, 16, 51, 13, 26, 21, 40, 37,
 59, 44]

plt.hist(no, color="g",edgecolor="w", cumulative=-1)

plt.title("Histogram")
plt.xlabel("python")
plt.ylabel("Number")

plt.show()
```

```python
no = [11, 29, 44, 55, 47, 28 ,58, 11, 57, 11, 17, 54, 27, 30, 59, 38, 36, 12, 46, 49, 10 ,34, 24, 42,
 35, 33, 51,14, 45, 22, 30 ,17, 38, 34, 36 ,24, 37, 48, 23, 37, 39, 16, 51, 13, 26, 21, 40, 37,
 59, 44]

plt.hist(no, color="g",edgecolor="b", histtype="stepfilled", orientation="horizontal")

plt.title("Histogram")
plt.xlabel("python")
plt.ylabel("Number")

plt.show()
```

```python
no = [11, 29, 44, 55, 47, 28 ,58, 11, 57, 11, 17, 54, 27, 30, 59, 38, 36, 12, 46, 49, 10 ,34, 24, 42,
 35, 33, 51,14, 45, 22, 30 ,17, 38, 34, 36 ,24, 37, 48, 23, 37, 39, 16, 51, 13, 26, 21, 40, 37,
 59, 44]

plt.hist(no, color="g",edgecolor="b", rwidth = 0.8, log=True, label="Python")

plt.title("Histogram")
plt.xlabel("python")
plt.ylabel("Number")

plt.axvline(45, color="m", label = "line")
plt.legend()
plt.grid()

plt.show()
```
