Scatter plot

```python
import matplotlib.pyplot as plt
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 2, 5]

c = ["r", "b", "g", "y", "r", "b", "g"]
s = [100, 200, 300, 450, 220, 340, 230]

plt.scatter(day, no, color = c, s = s, alpha=0.7, marker = "*", edgecolor = "m", linewidth=2)
plt.title("Scatter plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("Number", fontsize= 15)

plt.show()
```

```python
import matplotlib.pyplot as plt
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 2, 5]
no2 = [3, 5, 1, 6, 4, 8, 8]


c = [10, 29, 30, 40, 29, 46, 90]
s = [150, 200, 300, 450, 220, 340, 230]

plt.scatter(day, no, c = c, s = s, cmap="BrBG")

plt.scatter(day, no2, s= s, color="r")

t = plt.colorbar()
t.set_label("colorbar", fontsize = 15)
plt.title("Scatter plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("Number", fontsize= 15)

plt.show()
```
