```python
import matplotlib.pyplot as plt

x = ["python", "c", "c++", "javascript"]

y = [90, 52, 65, 85]
plt.xlabel("language", fontsize = 20)
plt.ylabel("popularity", fontsize = 20)
plt.title("First program", fontsize = 30)
c = ["y", "b", "g", "m"]
plt.bar(x,y, width = 0.4, color= c, edgecolor="r", linewidth=5, linestyle=":", alpha=0.8)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["python", "c", "c++", "javascript"]

y = [90, 52, 65, 85]
z = [20, 30, 40, 50]

p = np.arange(len(x))
width = 0.2
p1 = [j + width for j in p]

plt.xlabel("language", fontsize = 20)
plt.ylabel("popularity", fontsize = 20)
plt.title("First program", fontsize = 30)

plt.bar(p,y, width, color= "r",label="popularity")
plt.bar(p1,z, width, color= "g",label="download")

plt.xticks(p+width/2, x, rotation=20)
plt.legend()
plt.show()
```
