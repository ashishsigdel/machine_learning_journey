Piechart

```python
import matplotlib.pyplot as plt
x = [10, 20, 30, 40]
y = ["c", "c++", "java", "python"]

ex= [0.2, 0.0, 0.0, 0.0]
c = ["r", "b", "y", "g"]

plt.pie(x, labels=y, explode=ex, colors=c, autopct="%0.df%%", shadow=True, radius=1, labeldistance=1.1, startangle=180,
        textprops={"fontsize":15}, counterclock=False, wedgeprops={"linewidth" : 4}, rotatelabels = True)

plt.title("Pie Chart")

plt.legend(loc=2)

plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10, 20, 30, 40]
y = ["c", "c++", "java", "python"]
x1 = [40, 30, 20, 10]
c = ["r", "b", "y", "g"]

plt.pie(x, labels=y,radius = 1.5)
plt.pie(x1, radius=1, colors = c)

plt.title("Pie Chart")

plt.show()
```
